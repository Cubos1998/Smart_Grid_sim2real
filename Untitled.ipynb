```python
import pandas as pd
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import pandas as pd
import os
import numpy as np
import numpy as np
from matplotlib import pyplot as plt
from matplotlib.backends.backend_pdf import PdfPages
from pylab import *

matplotlib.rcParams['pdf.fonttype'] = 42
matplotlib.rcParams['ps.fonttype'] = 42
```

```python
def average_sliding_window(array,d):
    res = []
    t = 0
    while t < len(array):
        res.append(np.mean(array[t:t+d]))
        t += d
    return res
```

```python
fig = plt.figure(figsize=(8,5))

res0 = list(pd.read_csv('logfile/darc/'+'Humanoid_0901003-32.out',header = None,sep = 'shit')[0])

res = res0
source = []
target = []
tgt_100 = []
for t in res:
    
    
    if 'SOURCE' in t:
        source.append(float(t.split(":")[-1]))
    elif 'TARGET' in t:
        target.append(float(t.split(":")[-1]))
target = average_sliding_window(target,10)
source = average_sliding_window(source,10)

plt.plot(np.arange(len(target))*10,target,label = 'DARC reward on target domain')
plt.plot(np.arange(len(source)),source,label = 'DARC reward on source domain')
# plt.axhline(290)
# plt.axhline(390)
```
